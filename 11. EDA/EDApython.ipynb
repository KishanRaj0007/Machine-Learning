## Detecting Outliers
#### In statistics, an outlier is an observation point that is distant from other observations. The outliers can be a result of a mistake during data collection or it can be just an indication of variance in your data. If Ourliers are just a mistake, then we can remove it. But if it is a variance in data, we need to think more about this and we will also see about deciding whether a data is by mistake or just a variance.

```python
df.head()
```

```python
import seaborn as sns
sns.boxplot(x = df['HP'])
```

### Boxplot gives a five number summary of the data:- minimum, first quartile, median, third quartile, maximum. Herein all the plots, you can find some points are outside the box they are none other than outliers. The outlier detection and removing that I am going to perform is called IQR score technique

```python
from matplotlib import pyplot as plt
```

```python
plt.title('Identifying Outliers')
plt.boxplot(df['HP'])
plt.xlabel('HP Outliers')
plt.show()
```

```python
plt.title('Identifying Outliers')
plt.boxplot(df['HMPG'])
plt.xlabel('HMPG Outliers')
plt.show()
```

```python
plt.title('Identifying Outliers')
plt.boxplot(df['CMPG'])
plt.xlabel('CMPG Outliers')
plt.show()
```

```python
plt.title('Identifying Outliers')
plt.boxplot(df['MSRP'])
plt.xlabel('MSRP Outliers')
plt.show()
```

```python
df.shape
```

```python
Q1 = df[['HP', 'CMPG', 'HMPG', 'MSRP']].quantile(0.25)
Q3 = df[['HP', 'CMPG', 'HMPG', 'MSRP']].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
columns_to_filter = ['HP', 'CMPG', 'HMPG', 'MSRP']
filter_condition = ~((df[columns_to_filter] < (Q1 - 1.5 * IQR)) | (df[columns_to_filter] > (Q3 + 1.5 * IQR))).any(axis=1)
df = df[filter_condition]
```

```python
df.shape
```

```python
# Define the data and titles
columns = ['HP', 'CMPG', 'HMPG', 'MSRP']
titles = ['HP Outliers Result', 'HMPG Outliers Result', 'CMPG Outliers Result', 'MSRP Outliers Result']

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(15, 5))  # 1 row, 4 columns

# Loop through each column and plot the boxplot in a separate subplot
for i, col in enumerate(columns):
    axes[i].boxplot(df[col])
    axes[i].set_title(titles[i])
    axes[i].set_xlabel(col)

# Add a common y-label
fig.supylabel('Value')

# Display the plots
plt.tight_layout()  # Adjusts the subplot layout to avoid overlapping
plt.show()
```

## Scatter Plot
#### In the parameter of plt.scatter() function c is for setting color of points in scatter plot. Hence each unique year will correspond to different color in scatter plot. Also s parameter controls the size of points. Here df['Cylinders'] is multiplied by 50 to scale the size of points. Higher the size of cylinder larger the dots will appear on graph. Here alpha controls the transparency of the points. alpha = 0.6 means points are 60% opaque and 40% transparent.

```python
plt.figure(figsize = (10,6))
plt.scatter(df['HP'], df['MSRP'], c =df['Year'], s = df['Cylinders']*50, alpha = 0.6, cmap = 'viridis')
plt.colorbar(label = 'Year')
plt.xlabel('HP')
plt.ylabel('MSRP')
plt.title('HP vs MSRP (Size: Cylinders, Color: Year)')
plt.show()
```
