```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
plt.plot(df['year'], df['per capita income (US$)'], linewidth = 5)
plt.title('Year Wise income of US')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
x = df[['year']].values
y = df[['per capita income (US$)']].values
```

```python
model.fit(x,y) # 2d array expected as parameter
```

```python
ans = model.predict([[2020]])
ans
```

```python
years_to_predict = np.array([2020, 2021, 2022, 2023, 2024, 2025, 2026, 2027, 2028, 2029, 2030]).reshape(-1, 1)
#years_to_predict is reshaped to have each year as a separate sample with one feature per sample
ans = model.predict(years_to_predict)
ans
```

```python
plt.title("Prediction of yearly Income of US")
plt.xlabel("Years")
plt.ylabel("Income")
plt.plot(years_to_predict, ans)
plt.show()
```
